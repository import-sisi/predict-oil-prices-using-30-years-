```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('brent_oil_weekly_30_years.csv')
```

```python
df.info()
print(df.isna().sum())
```

```python
def convert_string_to_float(value):
    if isinstance(value, float):
        return value
    elif isinstance(value, str):
        if 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('K', '')) * 1e3
        elif '%' in value:
            return float(value.replace('%', ''))
    return float(value)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Vol.'] = df['Vol.'].fillna(0).apply(convert_string_to_float)
df['Change %'] = df['Change %'].apply(convert_string_to_float)
# numeric_columns = ['Price', 'Open', 'High', 'Low', 'Vol.', 'Change %']
# df[numeric_columns] = df[numeric_columns].interpolate()
print(df.isna().sum())
```

```python

print(new_df.isna().sum())
```

```python
n = 8
m = 1

new_df = create_time_series_dataframe(df, n, m)
```

```python
train_data = new_df.loc[new_df['Date'] < '2019-01-01']
test_data = new_df.loc[new_df['Date'] >= '2019-01-01']
```

```python
X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
y_train = train_data['Price_t+1']

X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
y_test = test_data['Price_t+1']
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tqdm import tqdm

models = [
    LinearRegression(),
    GradientBoostingRegressor(),
    SVR(),
    DecisionTreeRegressor()
]

for model in tqdm(models):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    print(f"{model.__class__.__name__}:")
    print(f"  MAE: {mae}")
    print(f"  MSE: {mse}")
```
