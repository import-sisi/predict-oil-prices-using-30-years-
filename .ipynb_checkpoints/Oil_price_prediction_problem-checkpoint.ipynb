```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from tqdm import tqdm
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('brent_oil_weekly_30_years.csv')
```

```python
def convert_string_to_float(value):
    if isinstance(value, float):
        return value
    elif isinstance(value, str):
        if 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('K', '')) * 1e3
        elif '%' in value:
            return float(value.replace('%', ''))
    return float(value)
```

```python
df.info()
print(df.isna().sum())
df['Date'] = pd.to_datetime(df['Date'])
df['Vol.'] = df['Vol.'].fillna(0).apply(convert_string_to_float)
df['Change %'] = df['Change %'].apply(convert_string_to_float) 
df.head()
```

```python

def create_time_series_dataframe(df, n, m):
    new_df = pd.DataFrame()
    print(len(df)-m)
    print(len(df))
    for i in range(n, len(df)-m):
        row = {}

        # Add Date column
        row['Date'] = df.loc[i, 'Date']

        # Add features from n weeks ago
        for j in range(1, n+1):
            row[f'Vol._t-{j}'] = df.loc[i-j, 'Vol.']
            row[f'Low_t-{j}'] = df.loc[i-j, 'Low']
            row[f'High_t-{j}'] = df.loc[i-j, 'High']
            row[f'Open_t-{j}'] = df.loc[i-j, 'Open']
            row[f'Price_t-{j}'] = df.loc[i-j, 'Price']
            row[f'Change %_-{j}'] = df.loc[i-j, 'Change %']

        # Add features for m weeks ahead
        for k in range(1, m+1):
            row[f'Price_t+{k}'] = df.loc[i+k, 'Price']

        new_df = pd.concat([new_df, pd.DataFrame(row, index=[0])], ignore_index=True)

    return new_df

# Example usage
n = 2
m = 2

new_df = create_time_series_dataframe(df, n, m)
print(new_df)
```

```python
train_data = new_df.loc[new_df['Date'] < '2019-01-01']
test_data = new_df.loc[new_df['Date'] >= '2019-01-01']
```

```python
X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
y_train = train_data['Price_t+1']

X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
y_test = test_data['Price_t+1']
y_train.head()
```

```python


models = [
    LinearRegression(),
    GradientBoostingRegressor(),
    SVR(),
    DecisionTreeRegressor()
]

for model in tqdm(models):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
     # Plot the graph
    plt.figure(figsize=(10, 6))
    plt.plot(y_test, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.xlabel('Steps')
    plt.ylabel('Price')
    plt.title(f'Prediction using {model.__class__.__name__}')
    plt.legend()
    plt.show()
    
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    print(f"{model.__class__.__name__}:")
    print(f"  MAE: {mae}")
    print(f"  MSE: {mse}")
```

```python
n_values = [2, 4, 6, 8, 10]  # Consider different values for n

results = []

for n in n_values:
    m = 1  # Set a constant value for m
    
    new_df = create_time_series_dataframe(df, n, m)

    train_data = new_df.loc[new_df['Date'] < '2019-01-01']
    test_data = new_df.loc[new_df['Date'] >= '2019-01-01']

    X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
    y_train = train_data['Price_t+1']

    X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
    y_test = test_data['Price_t+1']

    models = [
        LinearRegression(),
        GradientBoostingRegressor(),
        SVR(),
        DecisionTreeRegressor()
    ]

    model_results = {}

    for model in models:
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        mae = mean_absolute_error(y_test, predictions)
        mse = mean_squared_error(y_test, predictions)
        model_results[model.__class__.__name__] = {'MAE': mae, 'MSE': mse}

    results.append({'n': n, 'model_results': model_results})

# Display the results
for result in results:
    n = result['n']
    model_results = result['model_results']
    print(f"Results for n = {n}:")
    for model_name, metrics in model_results.items():
        mae = metrics['MAE']
        mse = metrics['MSE']
        print(f"{model_name}:")
        print(f"  MAE: {mae}")
        print(f"  MSE: {mse}")
    print('\n')
```

```python
import numpy as np
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

n = 8
m = 1

new_df = create_time_series_dataframe(df, n, m)

train_data = new_df.loc[new_df['Date'] < '2019-01-01']
test_data = new_df.loc[new_df['Date'] >= '2019-01-01']

X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
y_train = train_data['Price_t+1']

X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
y_test = test_data['Price_t+1']

feature_selector = SequentialFeatureSelector(LinearRegression(), n_features_to_select=12, direction='forward')
selected_features = feature_selector.fit_transform(X_train, y_train)

selected_feature_indices = feature_selector.get_support(indices=True)
selected_feature_names = X_train.columns[selected_feature_indices]

X_train_selected = X_train.iloc[:, selected_feature_indices]
X_test_selected = X_test.iloc[:, selected_feature_indices]

models = [
    LinearRegression(),
    # Add more models if needed
]

for model in models:
    model.fit(X_train_selected, y_train)
    predictions = model.predict(X_test_selected)
    
    # Plot the graph
    plt.figure(figsize=(10, 6))
    plt.plot(y_test, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.xlabel('Steps')
    plt.ylabel('Price')
    plt.title(f'Prediction using {model.__class__.__name__}')
    plt.legend()
    plt.show()
    
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    print(f"{model.__class__.__name__}:")
    print(f"  Selected Features: {', '.join(selected_feature_names)}")
    print(f"  MAE: {mae}")
    print(f"  MSE: {mse}")
```

```python
import itertools


n = 8
m = 1

new_df = create_time_series_dataframe(df, n, m)

train_data = new_df.loc[new_df['Date'] < '2019-01-01']
test_data = new_df.loc[new_df['Date'] >= '2019-01-01']

X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
y_train = train_data['Price_t+1']

X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
y_test = test_data['Price_t+1']

models = [
    LinearRegression(),
]

best_score = float('inf')
best_model = None
selected_features = []
k = 12

remaining_features = list(X_train.columns)

while len(selected_features) < k:
    best_score_iteration = float('inf')
    best_feature = None
    
    for feature in remaining_features:
        features_subset = selected_features + [feature]
        X_train_subset = X_train[features_subset]
        model = LinearRegression()
        model.fit(X_train_subset, y_train)
        X_test_subset = X_test[features_subset]
        predictions = model.predict(X_test_subset)
        
        mae = mean_absolute_error(y_test, predictions)
        mse = mean_squared_error(y_test, predictions)
        score = mse  # Use MSE as the performance metric
        
        if score < best_score_iteration:
            best_score_iteration = score
            best_feature = feature
            best_model = model
    
    selected_features.append(best_feature)
    remaining_features.remove(best_feature)
    best_score = best_score_iteration

    print(f"Selected Features: {selected_features}")
    print(f"Best Score: {best_score}")
    print(f"Best Model: {best_model.__class__.__name__}")
    print()
```

```python
from PyEMD import EMD
from PyEMD import EEMD

def add_emd_features(df):
   
    eemd = EEMD()

    price_series = df['Price'].values

    imfs = eemd.eemd(price_series)

    for i, imf in enumerate(imfs):
        df[f'IMF_{i+1}'] = imf

    return df


df_with_emd = add_emd_features(df)

print(df_with_emd)

plt.figure(figsize=(10, 6))

for i in range(1, 11):  
    plt.plot(df_with_emd['Date'], df_with_emd[f'IMF_{i}'], label=f'IMF_{i}')

plt.title('Decomposed IMFs')
plt.xlabel('Date')
plt.ylabel('IMF Values')
plt.legend()
plt.show()



plt.figure(figsize=(10, 6))

for i in range(1, 11):  
    plt.subplot(5, 2, i)  
    plt.plot(df_with_emd['Date'], df_with_emd[f'IMF_{i}'])
    plt.title(f'IMF_{i}')
    plt.xlabel('Date')
    plt.ylabel('IMF Values')


plt.tight_layout()

plt.show()
```

```python

n = 8
m = 1

new_df = create_time_series_dataframe(df, n, m)

train_data = new_df.loc[new_df['Date'] < '2019-01-01']
test_data = new_df.loc[new_df['Date'] >= '2019-01-01']

X_train = train_data.drop(['Date', 'Price_t+1'], axis=1)
y_train = train_data['Price_t+1']

X_test = test_data.drop(['Date', 'Price_t+1'], axis=1)
y_test = test_data['Price_t+1']

linear_regression = LinearRegression()
decision_tree_regression = DecisionTreeRegressor()
random_forest_regression = RandomForestRegressor()
svm_regression = SVR()

estimators = [
    ('linear_regression', linear_regression),
    ('decision_tree_regression', decision_tree_regression),
    ('random_forest_regression', random_forest_regression),
    ('svm_regression', svm_regression)
]
stacking_regressor = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())

stacking_regressor.fit(X_train, y_train)

predictions = stacking_regressor.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)

print(f"Stacking Regression MAE: {mae}")
print(f"Stacking Regression MSE: {mse}")
```
